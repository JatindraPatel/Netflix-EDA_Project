# Convert Date Time format

```python
df['date_added'] = df['date_added'].str.strip()
df["date_added"] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

# Relation between Type and Rating

```python
import plotly.express as px

rel = df.groupby(['type','rating']).size().reset_index(name='count')
fig = px.bar(rel, x='rating', y='count', color='type',
             title="Relation between Type and Rating",
             labels={'type':'Type','rating':'Rating','count':'Count'}).show()
```

# Top 10 Netflix Title Producing Directors

```python
import plotly.express as px
import matplotlib.pyplot as plt

top = df[df['director'] != 'Not Specified']['director'].value_counts().head(10).reset_index()
top.columns = ['director', 'count']
fig = px.bar(top, x='director', y='count', title="Top 10 Netflix Title Producing Directors",
             labels={'director':'Directors','count':'Count of Titles'})

fig.update_layout(title_font=dict(size=20, weight="bold"))
fig.update_xaxes(tickangle=-40)

fig.show()
```

# Top 10 Country Produncing Highest Netflix Titles

```python
import plotly.express as px
import matplotlib.pyplot as plt

top = df[df['country'] != 'Not Specified']['country'].value_counts().head(10).reset_index()
top.columns = ['country', 'count']
fig = px.bar(top, x='country', y='count', title="Top 10 Netflix Title Producing Countries",
             labels={'country':'Country','count':'Count of Titles'})
fig.update_layout(title_font=dict(size=20, weight="bold"))
fig.update_xaxes(tickangle=-40)

fig.show()
```

# Year Wise Trend

```python
import matplotlib.pyplot as plt
import plotly.express as px

df_year = df['release_year'].value_counts().sort_index().reset_index()
df_year.columns=['Year','Count']
fig = px.line(data_frame=df_year, x='Year', y='Count', title="Netflix Year-Wise Content Trend", markers=True)
fig.update_traces(marker_size=6)
fig.show()
```

# Movie vs Tv Show Growth Trend

```python
import plotly.express as px
import matplotlib.pyplot as plt


df['date_added'] = df['date_added'].dropna()
df['Added_Year'] = pd.to_datetime(df['date_added']).dt.year

df_type_year = df[df['country'] != 'Not Specified'].groupby(['Added_Year','type']).size().reset_index(name='Count')

fig = px.area(df_type_year, x='Added_Year', y='Count', color='type',
              title="Movies vs TV Shows Added Year-wise (Growth Trend)")
fig.show()
```

# Content Duration vs Release Year

```python
import plotly.express as px
import matplotlib.pyplot as plt


df_dur = df['duration'].dropna().str.extract('(\d+)').astype(float)
df['Duration_Num'] = df_dur

df_bubble = df.groupby(['release_year','Duration_Num']).size().reset_index(name='Count')

fig = px.scatter(df_bubble, x='release_year', y='Duration_Num', size='Count',
                 title="Release Year vs Content Duration Trend (Bubble Chart)",
                 labels={"release_year":"Release Year", "Duration_Num":"Duration (Minutes/Seasons)"})
fig.show()
```

# Month-Wise Content Added Pattern

```python
import plotly.express as px
import matplotlib.pyplot as plt

df['Added_Year'] = pd.to_datetime(df['date_added']).dt.year
df['Added_Month'] = pd.to_datetime(df['date_added']).dt.month_name()

df_heat = df.groupby(['Added_Year','Added_Month']).size().reset_index(name='Count')

fig = px.density_heatmap(df_heat, x='Added_Month', y='Added_Year', z='Count',
                         title="Month-Wise Netflix Content Addition Pattern",
                         labels={"Added_Month":"Month", "Added_Year":"Year"})
fig.show()
```

# Top 10 Actor`s

```python
import matplotlib.pyplot as plt
import plotly.express as px

actors = df[df['cast'] != 'Not Specified']['cast'].dropna().str.split(', ').explode()
top_actors = actors.value_counts().head(10).reset_index()
top_actors.columns = ['Actor','Count']
fig = px.bar(top_actors, x='Actor', y='Count', title="Top 10 Actors on Netflix ")
fig.update_xaxes(tickangle=-40)
fig.show()
```

# Netflix Content Preference by Age Group

```python
import matplotlib.pyplot as plt
import plotly.express as px

age_map = {
    "TV-Y": "Kids (0–6)",
    "TV-Y7": "Older Kids (7–12)",
    "TV-G": "Family (All Ages)",
    "G": "Family (All Ages)",
    "TV-G": "Family (All Ages)",
    "TV-PG": "Teens (13+)",
    "PG": "Teens (13+)",
    "PG-13": "Young Adults (16+)",
    "TV-14": "Young Adults (16+)",
    "TV-MA": "Adults (18+)",
    "R": "Adults (18+)",
    "NC-17": "Adults (18+)",
    "Not Specified": None
}

df['Age_Group'] = df['rating'].map(age_map)
df_age = df[df['Age_Group'].notna()].groupby(['type','Age_Group']).size().reset_index(name='Count')

fig = px.bar(df_age, x='Age_Group', y='Count', color='type', barmode='group',
             title="Netflix Content Preference by Age Group ", text='Count')
fig.update_traces(textposition='outside')
fig.show()
```
