# **DATA CLEANING** :-

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Not Specified')
df['rating'] = df['rating'].fillna('Not Specified')
df['duration'] = df['duration'].fillna('Not Specified')
df['country'] = df['country'].fillna('Not Specified')
df.dropna(subset=['date_added'], inplace=True)
df['cast'] = df['cast'].fillna('Not Specified')
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.strftime('%B')
df['added_day'] = df['date_added'].dt.day
```

```python
df['duration'] = df['duration'].str.replace(' min', '').str.replace(' Seasons', '').str.replace(' Season', '')
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')
df.rename(columns={'duration': 'duration_value'}, inplace=True)
```

```python
df.columns = [col.capitalize() for col in df.columns]
df.columns
```

```python
df.info()
```

```python
valid_ratings = [
    "TV-MA","TV-14","TV-PG","R","PG-13","TV-Y","TV-Y7","PG","TV-G","NR","G",
    "TV-Y7-FV","NC-17","UR","Not Specified"
]

df_clean = df[df['Rating'].isin(valid_ratings)]
```

```python
df.columns = [col.capitalize() for col in df.columns]
```

# **DATA - Visualization** :-

```python
# Count of movies and tv show by using pie chart...?

import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette("bright")
df['Type'].value_counts().plot(kind='pie', autopct='%2.1f%%', startangle=90, colors=colors)
plt.title('Count of Movies and TV Shows',fontweight='bold')
plt.ylabel('')
plt.show()
```

```python
# WAP to plot Top 10 Movies on Netflix using a bar chart...?

import matplotlib.pyplot as plt
import seaborn as sns

Movie = df[df['Type'] == 'Movie']['Title'].value_counts().head(10)
sns.barplot(x=Movie.index, y=Movie.values, palette="bright", hue=Movie.index, legend=False)
plt.title('Top 10 Movie on Netflix ', fontsize=18, fontweight='bold')
plt.xlabel('Movie Titles', fontsize=14)
plt.ylabel('Total Count', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.show()
```

```python
# WAP to plot Top 10 Tv show on Netflix using a bar chart...?
import matplotlib.pyplot as plt
import seaborn as sns

TV = df[df['Type'] == 'TV Show']['Title'].value_counts().head(10)

sns.barplot(x=TV.index, y=TV.values, palette="Set1", hue=TV.index, legend=False)
plt.title('Top 10 TV Shows on Netflix ', fontsize=18, fontweight='bold')
plt.xlabel('TV Show Titles', fontsize=14)
plt.ylabel('Total Count', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)


plt.show()
```

```python
# WAP to find out the Year-wise trend of content added using a line chart...?

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

year_counts = df['Release_year'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
plt.title('Year-wise Trend of Content Added', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Count', fontsize=14)


plt.show()
```

```python
# WAP to plot the distribution of content ratings (TV-MA, TV-14, PG, etc.) using a bar chart...?

import matplotlib.pyplot as plt
import seaborn as sns

df_clean['Rating'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Content Ratings', fontsize=16, fontweight='bold')
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```

```python
# WAP to plot Movies vs TV Shows Added per Year..?

import matplotlib.pyplot as plt
import seaborn as sns

df.groupby(['Added_year','Type']).size().unstack().plot(kind='line')
plt.title('Movie vs TV Shows Added Per Year', fontsize=16, fontweight='bold')
plt.show()
```

```python
# WAP to plot Movies vs TV Shows Added per Month..?

import matplotlib.pyplot as plt
import seaborn as sns

df.groupby(['Added_month','Type']).size().unstack().plot(kind='line')
plt.title('Movie vs TV Shows Added Per Month', fontsize=16, fontweight='bold')
plt.show()
```

```python
# WAP to plot Top 10 countries producing the highest number of Netflix titles using a bar chart.....?

import matplotlib.pyplot as plt
import seaborn as sns

top = df[df['Country'] != 'Not Specified']['Country'].value_counts().head(10).reset_index()
top.columns = ['Country', 'Count']
sns.barplot(data=top, x='Country', y='Count', palette="bright")

plt.title('Top 10 Netflix Title Producing Countries', fontsize=18, fontweight='bold')
plt.xlabel('Country', fontsize=14)
plt.ylabel('Count of Titles', fontsize=14)
plt.xticks(rotation=40, ha='right')
plt.show()
```

```python
# WAP to plot Top 10 directors producing the highest number of Netflix titles using a bar chart.....?

import matplotlib.pyplot as plt
import seaborn as sns

top = df[df['Director'] != 'Not Specified']['Director'].value_counts().head(10).reset_index()
top.columns = ['Director', 'Count']
sns.barplot(data=top, x='Director', y='Count', palette="bright")

plt.title('Top 10 Netflix Title Producing Directors', fontsize=18, fontweight='bold')
plt.xlabel('Directors', fontsize=14)
plt.ylabel('Count of Titles', fontsize=14)
plt.xticks(rotation=40, ha='right')
plt.show()
```
